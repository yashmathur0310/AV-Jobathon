```python
df['Discount']=df['Discount'].replace({'Yes':1,'No':0})
```

```python

df=df.drop(['ID'],axis=1)
df=pd.get_dummies(columns=['Store_Type','Location_Type','Region_Code'],data=df)
df['Day']=df['Date'].dt.day
df['Month']=df['Date'].dt.month
df=df.drop(['Date'],axis=1)
df=df.drop(['#Order'],axis=1)
```

```python
df
```

```python
x=df.drop(['Sales'],axis=1).values
y=df['Sales'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
model=keras.models.Sequential([
    keras.layers.Dense(45,input_shape=(18,),activation='relu'),
     keras.layers.Dense(40,input_shape=(18,),activation='relu'),
      keras.layers.Dense(30,input_shape=(18,),activation='relu'),
       keras.layers.Dense(20,input_shape=(18,),activation='relu'),
       keras.layers.Dense(10,input_shape=(18,),activation='relu'),
    keras.layers.Dense(1,activation='linear')])
model.compile(optimizer='adam',
             loss='mean_squared_error',
             )
model.fit(x_train,y_train,epochs=200,batch_size=512)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_log_error
msle=mean_squared_log_error(y_test,y_pred)
```
